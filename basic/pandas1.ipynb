```python
import numpy as np

num_rows = 30

roll_numbers = np.arange(1, num_rows + 1) 

names= ['Aryan', 'Aarav', 'Saanvi', 'Ishaan', 'Anaya', 'Vivaan', 'Aadhya', 'Reyansh', 'Kavya', 'Rohan',
                   'Pooja', 'Pratik', 'Meera', 'Rajesh', 'Aarti', 'Nisha', 'Ravi', 'Anita', 'Sanjay', 'Sheetal',
                   'Amit', 'Neeta', 'Nitin', 'Priti', 'Manish', 'Sunita', 'Kunal', 'Rita', 'Kiran', 'Suraj']
math_scores = np.random.randint(50, 101, size=num_rows)
science_scores = np.random.randint(50, 101, size=num_rows)
nepali_scores = np.random.randint(50, 101, size=num_rows)
english_scores = np.random.randint(50, 101, size=num_rows)

df = pd.DataFrame({
    'Name':names,
    'Roll': roll_numbers,
    'Math': math_scores,
    'Science': science_scores,
    'Nepali': nepali_scores,
    'English': english_scores
})

print(df)
df.info()
df.describe()
```
