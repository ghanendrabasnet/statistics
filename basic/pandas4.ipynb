```python
mi_matches = df[df['team1'] == 'Mumbai Indians']
mi_matches_sorted = mi_matches.sort_values('win_by_runs', ascending=False)

display(mi_matches_sorted)
```

```python
city_counts = df['city'].value_counts()
most_matches_city = city_counts.idxmax()  
most_matches_count = city_counts.max()   
print(f"The city that hosted the most matches is {most_matches_city} with {most_matches_count} matches.")
```

```python
matches_2017 = df[df['season'] == 2017]
selected_columns_2017 = matches_2017[['team1', 'team2', 'winner']]
display(selected_columns_2017)
```

```python
top_5_wickets = df.sort_values('win_by_wickets', ascending=False).head(5)
top_5_wickets_selected = top_5_wickets[['team1', 'team2', 'winner', 'win_by_wickets']]
display(top_5_wickets_selected)
```

```python
filtered_matches = df[(df['season'] == 20) & (df['win_by_wickets'] > 5)]
display(filtered_matches)
```

```python

filtered_matches = df[
    ((df['team1'] == 'Chennai Super Kings') | (df['team2'] == 'Chennai Super Kings')) &  
    (df['result'] != 'Tie') &  
    ((df['win_by_runs'] > 50) | (df['win_by_wickets'] > 5))  
]

display(filtered_matches)
```

```python
filtered_matches = df[
    ((df['team1'] == 'Kolkata Knight Riders') | (df['team1'] == 'Royal Challengers Bangalore')) &  
    (df['toss_decision'] == 'field') & 
    (df['win_by_runs'] < 20)  
]
display(filtered_matches)
```

```python
matches_with_winner = df[df['winner'].notna()]
selected_columns = matches_with_winner[['team1', 'team2', 'winner', 'win_by_wickets']]

sorted_matches = selected_columns.sort_values('win_by_wickets', ascending=False)
display(sorted_matches)
```

```python
matches_toss_winner_same = df[df['toss_winner'] == df['winner']]
player_of_match_counts = matches_toss_winner_same['player_of_match'].value_counts()
most_awards_player = player_of_match_counts.idxmax()
most_awards_count = player_of_match_counts.max()

print(f"The player with the most 'player_of_match' awards in matches where the toss_winner and winner are the same team is {most_awards_player} with {most_awards_count} awards.")
```
