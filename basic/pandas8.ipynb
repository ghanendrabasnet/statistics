```python
import pandas as pd
df=pd.read_excel(r'C:\\super.xlsx')
df
```

```python
##Calculate the total Sales and total Profit for each Region. Which region has the highest total profit?

region_sales_profit = df.groupby('Region').agg({'Sales': 'sum', 'Profit': 'sum'})

highest_profit_region = region_sales_profit['Profit'].idxmax()
highest_profit_value = region_sales_profit['Profit'].max()

print("Total Sales and Total Profit by Region:")
print(region_sales_profit)
print(f"\nRegion with the highest total profit: {highest_profit_region} with a profit of {highest_profit_value:.2f}")
```

```python
##Find the Sub-Category that has the highest total quantity sold.

sub_category_quantity = df.groupby('Sub-Category').agg({'Quantity': 'sum'})
highest_quantity_sub_category = sub_category_quantity['Quantity'].idxmax()
highest_quantity_value = sub_category_quantity['Quantity'].max()

print("Total Quantity Sold by Sub-Category:")
print(sub_category_quantity)
print(f"\nSub-Category with the highest total quantity sold: {highest_quantity_sub_category} with a quantity of {highest_quantity_value}")
```

```python
data={
    'sales':[500,200,300,100],
    "profit":[50,70,50,100]
}
df=pd.DataFrame(data)
df
```

```python
def profit_per(df):
    return df['profit']/df['sales']*100
df['per']=df.apply(profit_per,axis=1)
df
```

```python
data2={
    'roll':[1,2,3,4,5,6],
'Name':['ram','shyam','hemant','sita','gita','rita'],
    'per':[70,80,90,85,40,60]
}
df2=pd.DataFrame(data2)
df2
```

```python
def assign_grade(percentage):
    if percentage >= 90:
        return 'A+'
    elif percentage >= 80:
        return 'A'
    elif percentage >= 70:
        return 'B+'
    elif percentage >= 60:
        return 'B'
    else:
        return 'F'
```

```python
df2['Grade'] = df2['per'].apply(assign_grade)
df2
```
