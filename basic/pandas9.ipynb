```python
import pandas as pd
import numpy as np
data1={
    'emp_id':[101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'emp_name':["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah", "Ian", "Jane"],
    'post':["Manager", "Developer", "Analyst", "Developer", "Manager", "Analyst", "Developer", "Manager",
            "Analyst", "Developer"],
    'gender':["Female", "Male", "Male", "Male", "Female", "Male", "Female", "Female", "Male", "Female"],
    'salary':[85000, 75000, 65000, 70000, 90000, 68000, 72000, 88000, 66000, 74000]
}
df = pd.DataFrame(data1)
df
```

```python
df['salary'] = df['salary'] * 1.2
df
```

```python
df.loc[df['post'] == 'Developer', 'salary'] *= 1.2
(df)
```

```python
df.loc[(df['post'] == 'Developer') & (df['gender'] == 'Female'), 'salary'] *= 1.2
(df)
```

```python
def increase_salary(row):
    if row['post'] == 'Developer' and row['gender'] == 'Female':
        return row['salary'] * 1.2
    return row['salary']
df['salary'] = df.apply(increase_salary, axis=1)
(df)
```

```python
df['salary'] = df.apply(lambda row: row['salary'] * 1.2 if row['post'] == 'Developer' and row['gender'] == 'Female' else row['salary'], axis=1)
(df)
```
